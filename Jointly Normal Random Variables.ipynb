```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

Assume that there are two independent standard normal variables, U and V.
Using U and V, we create two random variables X and Y and they are jointly normal.

```python
mean = [1, 2]
cov = [[4, 0], [0, 3]]
```

```python
X, Y = np.random.multivariate_normal(mean, cov, 500).T
```

```python
plt.plot(X, Y, 'x')
plt.axis('equal')
plt.show()
```

Since X and Y are jointly normal, then any linear combination Z = aX + bY has a normal distribution.

```python
Z = 3*X + 5*Y
```

```python
sns.kdeplot(Z)
```

However, if X and Y are not jointly normal althought they have normal distribution, any linear combination of X and Y does not follow a normal distribution.

```python
X = np.random.normal(0, 1, 500)
```

The variable Y is constructed from X by flipping a coin; if the outcome is a tail (1), Y = X and Y = -X otherwise (0).

```python
samples = [random.randint(0, 1) for i in range(500) ]
heads = samples.count(0)
tails = samples.count(1)
Y = np.zeros(shape=(500,))
i = 0
for s in samples:
    if s == 0:
        Y[i] = X[i]
    else:
        Y[i] = -X[i]
    i += 1
```

```python
sns.kdeplot(Y)
```

We see that Y is also normally distributed. However, note that X and Y are not jointly normal. Now we create a new variable Z which is the sum of X and Y.

```python
Z = X + Y
```

```python
plt.plot(X, Y, 'x')
plt.axis('equal')
plt.show()
```

```python
sns.kdeplot(Z)
```

We easily can see that Z = X + Y is not normally distributed.
